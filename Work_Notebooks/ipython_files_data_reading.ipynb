```python
#–Њ—В–Ї—А—Л—В–Є–µ —Д–∞–є–ї–∞ –і–ї—П —З—В–µ–љ–Є—П —Б –њ–Њ–Љ–Њ—Й—М—О —Д—Г–љ–Ї—Ж–Є–Є open –Љ–Њ–і—Г–ї—П codecs —Б —Г–Ї–∞–Ј–∞–љ–Є–µ–Љ –Ї–Њ–і–Є—А–Њ–≤–Ї–Є koi8-r
file_obj = codecs.open('example_koi_8.txt', 'r', encoding='koi8-r')
```
